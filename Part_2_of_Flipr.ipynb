```
import math
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```
train_df = pd.read_excel("/content/time_series.xlsx")
```

```
train_df.rename(columns = {'Time Series Data':'April',"Unnamed: 2":"May","Unnamed: 3":"June","Unnamed: 4":"July","Unnamed: 5":"August"}, inplace = True)
train_df.drop(0)
train_df.head(50)
```

```
train_df.describe()
train_df["July"] = train_df["July"].fillna(train_df["August"])
train_df["June"] = train_df["June"].fillna(train_df["July"])
train_df["May"] = train_df["May"].fillna(train_df["June"])
train_df["April"] = train_df["April"].fillna(train_df["May"])
train_df.drop(0)
```

```
train_df.drop(columns="City",inplace = True)
train_df.drop(0 ,inplace = True)
```

```
train_df.apply(pd.to_numeric)
```

```
X = train_df.iloc[1:,:-1].values
y = train_df.iloc[1:,-1].values
```

```
X.astype('int64') 
y.astype("int64")
```

```
# X_train,X_test,y_train,y_test =  train_test_split(X,y,test_size=0.2,random_state = 27)
```

```
# X.info()
# X = X.apply(pd.to_numeric, errors='coerce')
```

```
X_test = train_df.iloc[:,1:]
X_test = X_test.values
X_test
```

```
from catboost import CatBoostRegressor
regressor = CatBoostRegressor(iterations=1200, depth=6, learning_rate=0.1, loss_function='RMSE',reg_lambda = 0.42,bagging_temperature = 100)
regressor.fit(X,y)
```

```
# y_pred_train = regressor.predict(X_train).astype(int)
y_pred = regressor.predict(X_test).astype(int)
y_pred.shape
```
